```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import dask.dataframe as dd
# use seaborn style for visuals
sns.set_theme()
```

```python
# read in clean_dataset.csv
df = pd.read_csv('Resources/Clean_Dataset.csv')
df.head()
#len(df)
```

```python
# read in economy.csv
economyDF = pd.read_csv('Resources/economy.csv')
economyDF['class'] = 'economy'
economyDF.head()
#len(economyDF)
```

```python
# read in business.csv
businessDF = pd.read_csv('Resources/business.csv')
businessDF['class'] = 'business'
businessDF.head()
#len(businessDF)
```

```python
# fdfd
unionAllDF = pd.concat([economyDF,businessDF])
unionAllDF.tail()
```

```python
unionAllDF['stop'].unique()
```

```python
# clean 'stop' column
def clean(x):
    if '1' in x:
        x = '1-stop'
    elif '2' in x:
        x = '2-stops'
    else:
        x = 'non-stop'
    return x

unionAllDF['stop'] = unionAllDF['stop'].apply(clean)
unionAllDF['stop'].apply(clean)
unionAllDF.head()
```

```python
# convert price to integer
unionAllDF['price'] = unionAllDF['price'].str.replace(',','')#.to_numeric(unionAllDF['price'])
unionAllDF['price'] = pd.to_numeric(unionAllDF['price'])
unionAllDF.head()
```

```python
# convert price from rupees to USD
unionAllDF['price'] = (unionAllDF['price'] *  0.01232)#.map('{:.2f}'.format) # add $ later if needed
unionAllDF.head()
```
